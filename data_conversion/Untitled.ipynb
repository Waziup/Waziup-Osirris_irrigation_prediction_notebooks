```python
import logging
import threading

logging.basicConfig(
    format="%(asctime)s - %(levelname)s - [%(threadName)s] - %(message)s",
    level=logging.INFO
)

def worker(name):
    logging.info(f"Starting work in {name}")
    for i in range(5):
        logging.info(f"Task {i} in {name}")
    logging.info(f"Finished work in {name}")

# Start multiple threads
thread1 = threading.Thread(target=worker, args=("Thread-1",), name="Worker-1")
thread2 = threading.Thread(target=worker, args=("Thread-2",), name="Worker-2")

thread1.start()
thread2.start()
thread1.join()
thread2.join()
```
