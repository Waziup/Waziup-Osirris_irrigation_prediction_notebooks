```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 100)

# Functions
y1 = 1 - x**2                 # Quadratic
y2 = 1 - x**1.5               # Generalized Power
y3 = np.exp(-.5*x) -.5              # Exponential Decay
y4 = np.cos(np.pi * x / 2)    # Cosine Decay

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label="1 - x (Quadratic)")
plt.plot(x, y2, label="1 - x^1.5 (Power, n=1.5)")
plt.plot(x, y3, label="e^(-x) (Exponential)")
plt.plot(x, y4, label="cos(x/2) (Cosine)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Comparison of Functions for Transition Weights")
plt.legend()
plt.grid()
plt.show()
```
